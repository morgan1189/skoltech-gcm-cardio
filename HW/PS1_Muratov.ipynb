## Problem 1) Resistor network problem. Stamping procedure.

In this problem we just need to demonstrate the stamping procedure for the node-branch type of problem. This requires reading the data from .mat file and forming a $ E $ matrix out of it, as well as $ I_s $ and $ \alpha $. This is done just by uploading the file provided along with the task formulation, extracting python matrices out of it (because originally it is in a form of a dictionary) and then construct named structures and use it using "solve" function from numpy package.

The code for the problem is below (**Note:** you need to have ps1_1_circuit.mat file in the same folder as this python notebook).

```python
import numpy as np
import scipy as sp
import scipy.io

mat = scipy.io.loadmat('ps1_1_circuit.mat')
Isource = mat['Isource']
R = mat['R']

Rsize = R.shape
Isourcesize = Isource.shape
nodes = max(R[1].max(), R[2].max())
branches = Rsize[1]

E = np.zeros((nodes, branches))
for i in range(branches):
    if (R[1][i] != 0):
        E[R[1][i]-1][i] += 1
    if (R[2][i] != 0):
        E[R[2][i]-1][i] -= 1

I = np.zeros(nodes)
for j in range(Isourcesize[1]):
    if (Isource[1][j] != 0):
        I[Isource[1][j]-1] = I[Isource[1][j]-1]-Isource[0][j]
    if (Isource[2][j] != 0):
        I[Isource[2][j]-1] = I[Isource[2][j]-1]+Isource[0][j]
        
alpha = np.zeros((branches, branches))
for k in range(Rsize[1]):
    alpha[k][k] = 1/float(R[0][k])
    
upperpart = np.concatenate((np.identity(branches), np.dot(-alpha, E.T)), axis=1)
lowerpart = np.concatenate((E, np.zeros((nodes, nodes))), axis=1)
nodebranchmat = np.r_[upperpart, lowerpart]

Ivector = np.r_[np.zeros(branches), I]
result = np.linalg.solve(nodebranchmat, Ivector)

currents = result[0:branches]
potentials = result[branches:branches+nodes]
```

## Problem 2) Heat conductivity problem.

<h3>a) Normal conditions</h3>
This problem has to do with the heat conductivity problem that was widely discussed in class. To solve it we have to perform a couple of steps:

* Discretize the $ x $ space
* Using a finite difference, approximate the result for the second derivative of T over x
* Separate the rod into discretized segment and write an equation for every piece
* Construct a matrix and a vector out of these expressions (taking into account the fact that the very first and the very least ones are maintained at temperatures $ T(0) $ and $ T(1) $ respectively)
* Solve a linear system

** 1) Discretizing the space. ** In my case I was just experimenting with different values until I obtained something feasible. We have to keep in mind that temperature along one segment should stay the same, because this is what is taken as for granted in solving the problem, thus having 1 or 2 segments is not really going to give you any meaningful result at the end (because we have the condition of keeping a constant temperature at both ends of the rod). A more elaborate answer to the very last question of the part a) of this problem would be — you must have enough segments so that your function and your derivative are continuous. In our case, all the functions were defined smooth enough so raising the number of segments just raises the precision of the answer.

** 2) Finite difference approximation. ** Here is how you can use finite difference to evaluate the second derivative. Let's take the segments $ i $, $ i-1 $ and $ i+1 $:

$$
\frac{\partial T(x)}{\partial x}|_{x_{i-\frac{1}{2}}} \approx \frac{T_i-T_{i-1}}{\Delta x}
$$

$$
\frac{\partial T(x)}{\partial x}|_{x_{i+\frac{1}{2}}} \approx \frac{T_{i+1}-T_{i}}{\Delta x}
$$

$$
\frac{\partial^2 T(x)}{\partial x^2}|_{x_{i}} \approx \frac{\frac{\partial T(x)}{\partial x}|_{x_{i+\frac{1}{2}}}-\frac{\partial T(x)}{\partial x}|_{x_{i-\frac{1}{2}}}}{\Delta x} \approx = \frac {T_{i-1}-2T_i+T_{i+1}}{\Delta x^2}
$$

** 3) Writing an updated equation for every segment. ** Having that in mind, are able to rewrite the equation we started from:

$$
\frac {1}{\Delta x^2}(T_{i-1}-2T_i+T_{i+1}) = \frac{\kappa_a}{\kappa_m(x_i)}
\left(T_i - T_0\right) - \frac{H(x_i)}{\kappa_m(x_i)}, x_i = x_0 + i\Delta x;
$$

$$
T_{i-1}\frac{1}{\Delta x^2}-T_i(\frac{2}{\Delta x^2}+\frac{\kappa_a}{\kappa_m(x_i)})+T_{i+1}\frac{1}{\Delta x^2} = - \frac{\kappa_a}{\kappa_m(x_i)}T_0 - \frac{H(x_i)}{\kappa_m(x_i)}
$$

** 4) Constructing a matrix. ** After all those steps the matrix has a really simple three-diagonal form, where two sub-diagonales correspond to $ i-1 $ and $ i+1 $ indices (this parts have $-\frac{1}{\Delta x^2}$). The diagonal part has $ \frac{2}{\Delta x^2}-\frac{\kappa_a}{\kappa_m(x_i)} $. This looks pretty straight-forward until we realize that we don't have a $ i-1 $ piece for the very first segment as well as $ i+1 $ piece for the very last segment. This is where the right part of the equation comes in handy.

Most elements of this vector look like this: $ - \frac{\kappa_a}{\kappa_m(x_i)}T_0 - \frac{H(x_i)}{\kappa_m(x_i)} $. The only difference lies in the first and last segments where we substract a constant of $T(0)$ (or $T(1)$ respectively) divided by $\Delta x^2$. Such a construction models a presence of an additional segment to the left (or to the right respectively) of the sides of the rod with a constant given temperature, which is used as $i-1$ (or $i+1$) segment for the bounding segments. These special virtual segments do not exchange heat with either air or heating source and are introduced just to ensure that initial conditions are satisfied.

The matrix looks something like this:

$$ \begin{bmatrix}
-(\frac{2}{\Delta x^2}+\frac{\kappa_a}{\kappa_m(x_i)}) & \frac{1}{\Delta x^2} & ... & ... & 0  \\ 
\frac{1}{\Delta x^2} & -(\frac{2}{\Delta x^2}+\frac{\kappa_a}{\kappa_m(x_i)}) & \frac{1}{\Delta x^2} & ... & 0 \\
\\ 0 & \frac{1}{\Delta x^2} & -(\frac{2}{\Delta x^2}+\frac{\kappa_a}{\kappa_m(x_i)}) & \frac{1}{\Delta x^2} & 0\\
\\ ... & ... & ... & ... & ... \\
\\ 0 & ... & ... & \frac{1}{\Delta x^2} & -(\frac{2}{\Delta x^2}+\frac{\kappa_a}{\kappa_m(x_i)}) \\
\end{bmatrix} \begin{bmatrix}
T_1  \\
... \\
T_{discr}  \\ 
\end{bmatrix} = \begin{bmatrix}
-\frac{T(0)}{\Delta x^2}-T_0\frac{\kappa_a}{\kappa_m(0)}-\frac{H(0)}{\kappa_m(0)}  \\
-T_0\frac{\kappa_a}{\kappa_m(\Delta x)}-\frac{H(\Delta x)}{\kappa_m(\Delta x)} \\
-T_0\frac{\kappa_a}{\kappa_m(2\Delta x)}-\frac{H(2\Delta x)}{\kappa_m(2\Delta x)} \\
... \\
-\frac{T(1)}{\Delta x^2}-T_0\frac{\kappa_a}{\kappa_m(1)}-\frac{H(1)}{\kappa_m(1)} \\ 
\end{bmatrix} $$

** 5) Solving a linear system. ** The implementation of all these steps as well as a graph of temperature along the length of the rod are presented below.

```python
import numpy as np
import scipy as sp
import scipy.io
import math
import matplotlib.pyplot as plt

def Heat(x):
    return 35*math.pow(math.cos(2*math.pi*float(x)),2)

def Heatproblem(discr):
    Ka = 0.005
    Km = 0.1
    Tair = 250
    T0 = 200
    T1 = 200

    delta = 1/float(discr)
    matrix = np.zeros((discr-1, discr-1))
    vector = np.zeros(discr-1)
    vector[0] = -T0/math.pow(float(delta),2)-(Tair*Ka/float(Km))-Heat(0)/float(Km)
    vector[discr-2] = -T1/math.pow(float(delta),2)-(Tair*Ka/float(Km))-Heat(1)/float(Km)
    for i in range(discr-1):
        if (i-1 >= 0):
            matrix[i][i-1] = 1/math.pow(float(delta),2)
        matrix[i][i] = -2/math.pow(float(delta), 2) - Ka/float(Km)
        if (i+1 <= discr-2):
            matrix[i][i+1] = 1/math.pow(float(delta),2)
        if ((i > 0) and (i < discr-2)):
            vector[i] = -(Tair*Ka/float(Km))-Heat(i*delta)/float(Km)

    T = np.linalg.solve(matrix, vector)
    realT = np.r_[T0, T]
    realT = np.r_[realT, T1]
        
    return realT

def Plotheatproblem(discr):
    delta = 1/float(discr)
    x = np.zeros(discr+1)
    for j in range(discr+1):
        x[j] = j*delta
    realT = Heatproblem(discr)
    
    %matplotlib inline
    plt.plot(x, realT)
    plottitle = 'Temperature distribution: %d segments' %(discr)        
    plt.title(plottitle)
    plt.xlabel('Length')
    plt.ylabel('Temperature')
    plt.show()
    print('Maximum temperature value: %f' %(max(realT)))
    return
    
Plotheatproblem(200)
```

<h3>b) Special conditions</h3>
I would say that obtaining a solution without having information about $ T_0 $ and $ T_1 $ is impossible because of the fact that we are not limited by enough conditions. This leads to the indefinite number of solutions which is caused or described by the fact that the conductivity matrix becomes singular. 

First, we can't anymore use finite difference method in the same way as we did before, because we must state that $ \frac{\partial T(x)}{\partial x} = 0 $. This results in having -1 1 in the first row and 1 -1 in the last row (instead of -2 1 and 1 -2 as before). 

Second, given the conditions of $K_a = 0$ and $H(x) = const$ lead to having the constant vector in the right part of the equation (and this vector also does not include neither $T(0)$ nor $T(1)$).

All of these factors contribute to having singular system and an infinite number of solutions hence it's impossible to solve the problem under these conditions.

Though representing a single neuron as a node and axons and dendrites as branches in such a system seems to be a no-brainer, this is actually not exactly true. Neuron is certainly a component, but not just a single node. We have to take into account the fact that the signal transmitting through axons and dendrites has certain properties:

* It's amplitude changes during transmission (signal weakening)
* The neuron potential declines with time
* The neuron fires only when it's potential reaches a certain level

We are still undecided whether we should try to stay as close to the proposed model as possible, or to simplify a few things but still we're able to answer the basic questions:

** What do the nodes of your network represent? **
They represent pre-synapstic and post-synaptic compartments. We may also discretize axons or dendrites to model the potential loss over distance.

** What are the quantities associated with the nodes? **
Voltages.

** What do the edges of your network represent? **
Edges represent any kind of structures that transform signal in any sense. There will be edges modelling the loss of potential over time and over distance (dendrites and axons) and there will be edges for the synapses.

** What are the quantities associated with the edges? **
Currents (or ion concentration, if that's more applicable in terms of the computational complexity and desired accuracy).

** What are the conditions that specify which nodes are connected by edges? ** 
The chosen topology of the network. We will probably have voltage sources which will excite the whole network (controlled input). We will also try to model the whole network having only one kind of synapses first. The idea is to start with simpler networks (less than 10 neurons and 50 synapses) and try to scale up if possible.

** What are the conditions that specify the behavior of the edges? **
For this part we will try to simplify existing equations proposed by Izhikevich:

$$
C\dot{v} = k(v-v_r)(v-v_t)-u+I
$$
$$
\dot{u} = a(b(v-v_t)-u)
$$

where $C$ is the membrane capacitance, $v$ is the membrane potential, $v_r$ is the resting potential, $v_t$ is the threshold potential, $u$ is the recovery variable (the difference of all inward and outward voltage-gated currents), $I$ is the dendritic and synaptic current summed up:

$$
I = -I_{dendr}-I_{syn}
$$

Implementing this model in terms of node-branch approach or nodal analysis may require additional discretization and introducing other nodes.
